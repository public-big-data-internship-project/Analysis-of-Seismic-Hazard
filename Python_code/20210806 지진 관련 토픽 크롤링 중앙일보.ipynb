### 중앙일보

```python
import pandas as pd
from selenium.webdriver import Chrome
from selenium import webdriver

print("검색")
search=input()

news=[]

options = webdriver.ChromeOptions()
options.add_argument('--window-size=800x600') # 창 크기 조절

browser = webdriver.Chrome('chromedriver', options=options) # 크롬 브라우저 켜짐
fronturl ='https://news.joins.com/Search/JoongangNews?Keyword='
backurl='&SortType=New&SearchCategoryType=JoongangNews&PeriodType=All&ScopeType=All&ImageType=All&JplusType=All&BlogType=All&ImageSearchType=Image&TotalCount=0&StartCount=0&IsChosung=False&IssueCategoryType=All&IsDuplicate=True&Page=1&PageSize=10&IsNeedTotalCount=True'
browser.get(fronturl+search+backurl) # url 이동
browser.implicitly_wait(1) # 기다려

for page in range(1,11):
    for _news in range(1, 11):             
        browser.find_element_by_xpath('//*[@id="content"]/div[2]/div[2]/ul/li['+str(_news)+']/div/h2/a').click() # 뉴스 클릭
        browser.implicitly_wait(1) # 기다려 
        browser.switch_to.window(browser.window_handles[1]) # 클릭한 뉴스 창 위치로 변경
        browser.implicitly_wait(0.5) # 기다려
        headline=browser.find_element_by_xpath('//*[@id="article_body"]').text
        news.append(headline)
        browser.close()
        browser.switch_to.window(browser.window_handles[0]) # 변경하기 전 위치로 변경
    browser.find_element_by_xpath('//*[@id="content"]/div[2]/div[3]/div/a['+str(page)+']').click() # 다음 페이지
browser.quit()
```
