　　└해당 코드는 위도와 경도를 수집했다는 과정을 보여주기 위한 예시 코드임

```python
import googlemaps

addr=[]
_lat=[]
_lng=[]
count=1

gmaps=googlemaps.Client(key='MY-API-KEY')

for x in range(0, len(read_gangju_build['대지위치'])):
    try:
        tmp=gmaps.geocode(read_gangju_build['대지위치'][x], language='ko')
        tmp_loc=tmp[0].get('geometry')
        _lat.append(tmp_loc['location']['lat'])
        _lng.append(tmp_loc['location']['lng'])
        print(count, _lat, _lng)
        count+=1
    except IndexError:
        _lat.append(0)
        _lng.append(0)
        print(count, _lat, _lng)
        count+=1
        
```

```python
read_gangju_build.loc[:,'위도'] = _lat
read_gangju_build.loc[:,'경도'] = _lng    
read_gangju_build
```
